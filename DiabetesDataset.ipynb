```python
#Check if any of data is null
b = pd.isnull(df)
print(" Null values in data:", b)
sns.countplot(data=df.fillna('Null', inplace=True))
```

```python
#from the description statistics, box plot and histogram, we could see that the columns Glucose, BloodPressure, 
#SkinThickness, Insulin, BMI have 0 as value which is invalid and seems like an outlier as the valves of these columns 
#ideally be never 0. let's calculate the outliers
#Replace value 0 with suitable values
#Step1 : Replace 0 with nan
#step2 : replace nan with suitable value

import numpy as np
C = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']

df = df.copy(deep = True)
df[C] = df[C].replace(0,np.NaN)

## showing the count of Nans
print(df.isnull().sum())

```

```python
#Based on the histogram, we can replace the nan values for those columns with median 

df['Glucose'].fillna(df['Glucose'].median(), inplace = True)
df['BloodPressure'].fillna(df['BloodPressure'].median(), inplace = True)
df['SkinThickness'].fillna(df['SkinThickness'].median(), inplace = True)
df['Insulin'].fillna(df['Insulin'].median(), inplace = True)
df['BMI'].fillna(df['BMI'].median(), inplace = True)

df.describe()
```

```python
#Histogram after replacing 0 with median

df.hist(bins=30, figsize=(20,20),color='darkblue')
plt.show()
```

```python
#Data Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns=['Pregnancies', 'Glucose','BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
df[columns] = scaler.fit_transform(df[columns])
df.info()
df.head()
```

```python
correlations = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True)
```

```python
correlations.Outcome.sort_values(ascending=False)
```

```python
#feature selection using Correlation. Bucket the features which have correlation greater than 0.8 
#with the target variable Outcome

correlated_features = []
for i in range(len(correlations.columns)):
    for j in range(i):
        if abs(correlations.iloc[i, j]) > 0.8:
            colname = correlations.columns[i]
            correlated_features.add(colname)
print(correlated_features)
print(len(correlated_features))
```

```python
#Since no feature has correlation greater tyhan 0.8, considering all input variables as fatures
X = df.iloc[:, 0:8] #independent variables
Y = df.iloc[:, -1] # target variable
print("Feature variables contains:", X.shape)
print("Target variable contains:", Y.shape)
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
#split data into train data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=50, stratify = Y)
print('shape of class X train data: ')
print(X_train.shape)
print('shape of class Y train data: ')
print(Y_train.shape)
print('shape of class X test data: ')
print(X_test.shape)
print('shape of class Y test data: ')
print(Y_test.shape)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn import metrics
scores = {'Model':[], 'Accuracy':[], 'CV_Score':[], 'MAE':[], 'MSE':[], 'RMSE':[]}
```

```python
#Model -1 : Decision Tree with default criteria and sample leaf
from sklearn.tree._classes import DecisionTreeClassifier

desitree = DecisionTreeClassifier()
desitree.fit(X_train, Y_train) # model with train data
predictions = desitree.predict(X_test) # predicting using test data

print("DecisionTreeclassifier Model")
#classification report for Decision Tree
cf=classification_report(Y_test, predictions)
print("classification report:")
print(cf)
# Use score method to get accuracy of model
Acc_score = accuracy_score(Y_test, predictions)
print("Accuracy Score: ", (Acc_score*100))
cv_score = np.mean(cross_val_score(desitree, X_test, Y_test, cv=5))
print("CV_score:", cv_score*100)
cm = confusion_matrix(Y_test, predictions)
print("confusion matrix:")
print(cm)

#Regression evaluation metrics
mae = metrics.mean_absolute_error(Y_test,predictions)
mse = metrics.mean_squared_error(Y_test,predictions)
rmse = np.sqrt(metrics.mean_squared_error(Y_test,predictions))
print("Regression evaluation metrics:")
print( 'MAE [ ', mae , ' ] MSE [ ', mse ,  ' ] RMSE [ ', rmse , ' ]')    
     
#To plot confusion matrix of Decision Tree
plt.figure(figsize=(15,15))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
plt.title(f"Confusion Matrix for DecisionTreeclassifier Model-Accuracy Score:" +str("{:.2f}".format(Acc_score*100)), 
         size = 15);
plt.show()     

scores['Model'].append('Decision Tree')
scores['Accuracy'].append(Acc_score*100)
scores['CV_Score'].append(cv_score*100) 
scores['MAE'].append(mae) 
scores['MSE'].append(mse)
scores['RMSE'].append(rmse)
```

```python
print("Predicted Data using Decision Tree model:")
print(predictions)
```
